```python
import pandas as pd

import warnings
warnings.filterwarnings(action='ignore')
```

```python
df1 = pd.read_csv('농산물.csv', index_col= 0).iloc[1:,:]
row_index = []
for i in range(0,len(df1), 2):
    row_index.append(i)

df = df1.iloc[row_index,:]

df.reset_index(inplace=True)

df = df.rename(columns={'index':'년도'})
df['년도'] = df['년도'].str[:4]

year = []
for i in range(len(df.columns)-1):
    for j in range(len(df['년도'])):
        year.append(df['년도'][j])
len(year)

df_dict = {}
df_dict['년도'] = year
item = []
for i in range(1,len(df.columns)):
    for j in range(len(df)):
        item.append(df.columns[i])
df_dict['품목'] = item

price = []
for col in df.columns[1:]:
    for idx in range(len(df[col])):
        price.append(df[col][idx])
df_dict['수출량'] = price

df_ = pd.DataFrame(df_dict)

df_.to_excel('농산물.xls')
```
